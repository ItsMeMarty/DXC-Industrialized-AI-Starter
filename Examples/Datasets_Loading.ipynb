Finding right or sample data set is one of the important tasks for those who are starting to work on Machine learning and AI. Though we have many datesets readily available finding right dataset, loading and using it can always be challanging specically for beginners. DXC Industrialized AI Starter make this easy by providing few data sets and function to load them easily without any difficulties.

Below code sinpet helps you to load and view the datasets from this library.

Below function helps in loading the dataset from AI Starter Library. This function have three options and explained below.

1. **ai.get_data('iris')**: It takes filename as input and loads that particular datset into your notebook and return as a pandas dataframe.

2. **ai.get_data('iris', display_details=True)**: Along with loading the dataset, display_details option displays details about the dataset. This parameter default will be set to False.

3. **ai.get_data('iris', display_details=True, save_copy=True)**: Along with loading dataset and printing details, save_copy options saves a copy of dataset in csv format in the same folder when notebook is used. This parameter default will be set to False.
